# Distributions

## Normal distribution

In first example i will demonstrate how to make a normal distribuition in python: 

here $ \mu = 0.5 $ and $ \sigma = 0.1$ the formula is defined as:

$  f(x) = \frac{1} {\sigma\sqrt{2\pi}} e^{-(x - \mu)^{2}/(2\sigma^{2}) }  $


```python
import matplotlib.pyplot as plt
import numpy as np

mu, sigma = 0.5, 0.1
s = np.random.normal(mu, sigma, 10000)

count, bins, ignored = plt.hist(s, 20, normed=True)

f_x = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2))

plt.plot(bins, f_x, linewidth=3, color='y')
plt.show()
```

# Binomial distribution



```python
from scipy.stats import binom
import seaborn as sb

s = binom.rvs(n=20,p=0.8,loc=0,size=1000)


ax = sb.distplot(s, kde=True, color='blue', hist_kws={"linewidth": 25,'alpha':1})
ax.set(xlabel='Binomial', ylabel='Frequency')
```

## Poisson distribution

```python
from scipy.stats import poisson
import seaborn as sb

data_binom = poisson.rvs(mu=4, size=10000)
ax = sb.distplot(data_binom, kde=True, color='green', hist_kws={"linewidth": 25,'alpha':1})
ax.set(xlabel='Poisson', ylabel='Frequency')
```

# Chi Squared Distribution

Elevating the Students T distribution to squared creates Chi Squared Distribution.

if $Z$ are values of normal distribution so...

$$\chi^2 =  \sum_{i = 1}^{k} {Z_i^2}$$

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

ax = sns.distplot(random.chisquare(df=1, size=1000), hist=False)
ax.set(xlabel='Chi', ylabel='Frequency')
```

# Logistic distribution

This distribution is good for accuracy distributions.
They have two variables for location and scale.

$ f(x; \mu,s) = \frac{e^{-(x-\mu)/s}} {s\left(1+e^{-(x-\mu)/s}\right)^2} $

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.logistic(loc=1,scale= 2,size=10), hist=False)

plt.show()
```

# How to calculate mean, mediam and mode in python?

For calculate we gona use statistics library.
### Mean:

its simple formula $ \sum_{i = 1}^{k} {X_i} / n$ here n is number of values. For the aritmethic mean (Exist other types of means)

```python
import statistics

statistics.mean([4, 8, 6, 5, 3, 2, 8, 9, 2, 5])
```

### Mediam:

Is the value that occur in half of the sorted list. 

```python
statistics.median([4, 8, 6, 5, 3, 2, 8, 9, 2, 5])
```

### Mode:

Is the value with most occurences.

```python
statistics.mode([4, 8, 6, 5, 2, 2, 8, 9, 2, 5])
```
